```python
# Importando as bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Pré-processamento de dados

dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
# a função apriori não aceita o formado dataframe da função read_csv
# tem que transformar o dataset em uma list

transactions = []
for i in range(0, 7501):
  transactions.append([str(dataset.values[i, j])for j in range(0, 20)])
```

```python
# Treinando o modelo Apriori

from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
# Visualizando os resultados em um dataframe organizado

results = list(rules)
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

```python
# Mostrando os resultados não-ordenados

resultsinDataFrame
```

```python
# Mostrando os resultados ordenados

resultsinDataFrame.nlargest(n = 10, columns = 'Lift')
```
